```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
# Load dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
# Gabungkan fitur dan label untuk korelasi
df = X.copy()
df['Diagnosis'] = y
# Hitung korelasi
correlation = df.corr()['Diagnosis'].abs().sort_values(ascending=False)
# Visualisasi
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('Heatmap Korelasi Fitur dengan Diagnosis')
plt.show()
# Tampilkan fitur berkorelasi tinggi
print(correlation)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# Gunakan 20 fitur terpilih
top_features = correlation.index[1:21]  # skip 'Diagnosis' itu sendiri
X_selected = X[top_features]
# Inisialisasi model
model = LogisticRegression(max_iter=10000)
# RFE untuk memilih 10 fitur
rfe = RFE(estimator=model, n_features_to_select=10)
fit = rfe.fit(X_selected, y)
# Hasil fitur terbaik
selected_features = X_selected.columns[fit.support_]
print("Fitur Terpilih:", list(selected_features))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
# Split data
X_train, X_test, y_train, y_test = train_test_split(X[selected_features],
y, test_size=0.3, random_state=42)
# Train model
model.fit(X_train, y_train)
# Prediksi
y_pred = model.predict(X_test)
# Evaluasi
print(classification_report(y_test, y_pred))
```
